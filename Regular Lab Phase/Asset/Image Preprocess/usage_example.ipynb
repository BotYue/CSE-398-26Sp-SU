```python
from PIL import Image, ImageOps

def preprocess_and_save(input_path, output_path, invert=False):
    img = Image.open(input_path).convert("L")
    if invert:
        img = ImageOps.invert(img)
    img = img.resize((28, 28), Image.BILINEAR)
    img.save(output_path)

    print("Saved processed image to:", output_path)

preprocess_and_save(
    input_path="CoachBag.png",
    output_path="CoachBag_28x28.png",
    invert=True  # Invert BW color
)
```

```python
from IPython.display import display
img = Image.open("CoachBag.png")
display(img)
```

```python
img = Image.open("CoachBag_28x28.png")
display(img)
```
